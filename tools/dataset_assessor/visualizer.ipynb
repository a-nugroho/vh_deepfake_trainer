```python
from PIL import Image
import matplotlib.pyplot as plt
from PIL import Image
import os

def plot_images_grid(image_paths, rows=6, cols=4, figsize=(18, 12),names=None):
    """
    Plot images in a grid layout using PIL for loading.

    Args:
        image_paths (list): List of image file paths.
        rows (int): Number of rows in grid.
        cols (int): Number of columns in grid.
        figsize (tuple): Figure size in inches.
    """
    fig, axes = plt.subplots(rows, cols, figsize=figsize)
    axes = axes.flatten()
    if names is None:
        names = [f"Image {i}" for i in range(len(image_paths))]

    for id_img,(ax, img_path) in enumerate(zip(axes, image_paths)):
        if os.path.exists(img_path):
            img = Image.open(img_path)
            ax.imshow(img)
            ax.set_title(names[id_img], fontsize=8)
            #ax.set_title(os.path.basename(img_path), fontsize=8)
        ax.axis("off")

    # Hide unused axes
    for ax in axes[len(image_paths):]:
        ax.axis("off")

    plt.tight_layout()
    plt.show()
```

```python
# Example usage
#folder_dir = "/mnt/ssd/datasets/deepfake/vh_55plus/raw_images/selfie"
#folder_dir = "/mnt/NAS/dataset/deepfake/vh_deepfake/ghostface/processed/"
folder_dir = "/mnt/NAS/dataset/deepfake/vh_deepfake/REFace/processed/"
list_files = sorted(os.listdir(folder_dir))
image_files = [os.path.join(folder_dir,i) for i in list_files[24:]]
plot_images_grid(image_files, rows=2, cols=6, figsize=(18, 6))
```
