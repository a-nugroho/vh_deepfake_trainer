```python
import pandas as pd
```

```python
def get_mean(df):
    counts = df.groupby('pred_label').size()
    ratios = counts / len(df)
    result = pd.concat([counts, ratios], axis=1, keys=['count', 'ratio'])
    return result
```

```python
csv_file = "test_model_output-bfs.txt"
df = pd.read_csv(csv_file,sep='|')
get_mean(df)
```
